```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

# Pipeline with increased max_iter and early stopping
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Scale data
    ('logreg', LogisticRegression(max_iter=5000, tol=1e-4))  # Early stopping
])

# Update parameter grid
param_grid = {
    'logreg__penalty': ['l1', 'l2'],
    'logreg__C': [0.001, 0.009, 0.01, 0.09, 1, 5, 10, 25, 50, 100],
    'logreg__solver': ['liblinear', 'saga']  # Solvers for l1 and l2
}

# Perform grid search
grid_clf_acc = GridSearchCV(pipeline, param_grid=param_grid, scoring='accuracy', cv=5)
grid_clf_acc.fit(X_train, y_train)

# Predict and display results
y_pred = grid_clf_acc.predict(X_test)
print("Best Parameters:", grid_clf_acc.best_params_)
print("Best Accuracy:", grid_clf_acc.best_score_)
```
